```python
map3=folium.Map(location=[19,72],zoom_start=6,tiles=" Stamen Terrain")
fg=folium.FeatureGroup('My map')
lat=list(data['LAT'])
lon=list(data['LON'])
el=list(data['ELEV'])
for x,y,z in zip(lat,lon,el):
    if z<2000:
        col='green'
    elif z>2000 and z<3000:
        col='orange'
    else:
        col='red'
    #zip is used to combine when we iterate through 2 objects
    fg.add_child(folium.Marker(location=[x,y],popup=str(z)+' m', icon=folium.Icon(color=col)))
map3.add_child(fg)
map3.save('Multiple-Marker-Mumbai.html')
```

```python
map5=folium.Map(location=[19,72],zoom_start=6,tiles=" Stamen Terrain")
fg1=folium.FeatureGroup('Volcanoes')
lat=list(data['LAT'])
lon=list(data['LON'])
el=list(data['ELEV'])
for x,y,z in zip(lat,lon,el):
    if z<2000:
        col='green'
    elif z>2000 and z<3000:
        col='orange'
    else:
        col='red'
    #zip is used to combine when we iterate through 2 objects
    fg1.add_child(folium.CircleMarker(location=[x,y],popup=str(z)+' m', fill_color=col,color='black',fill_opacity=1))
    
fg2=folium.FeatureGroup('Population Density')
fg2.add_child(folium.GeoJson(data=open('world.json','r',encoding='utf-8-sig').read(),
                            style_function=lambda x: {'fillColor':'green' if x['properties']['POP2005'] < 100000000 else 'red'} ))
map5.add_child(fg1)
map5.add_child(fg2)

# recent changes made are we separated the FeatureGroups because then we get an option to control the layers
# we also added folium.LayerControl() as a child class to the map.


map5.add_child(folium.LayerControl())


map5.save('Multiple-Circle-Marker-Mumbai-with-data-json.html')
# here we have used CircleMarker.
# it doesn't have a property for icon, but has a property of fill_color and color= 
# fill color, colour's it and color is the boundary color
```
