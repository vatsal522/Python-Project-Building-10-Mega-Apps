```python
import bokeh

from bokeh.plotting import figure
from bokeh.io import output_file,show

#prepare data
x=list(range(1,6))
y=list(range(6,11))
z=list(range(11,16))
#number of values in x and y are to be same, as they're mapped to each other


#prepare OUTPUT FILE
output_file('Line.html')

#create figure object
f=figure()

#create line plot
f.line(x,y)
f.triangle(x,y,z)
f.circle(x,z)
show(f)

```

```python
import bokeh
import pandas
from bokeh.plotting import figure
from bokeh.io import output_file,show

#prepare data
df=pandas.read_csv('bachelors.csv')
x=df['Year']
y=df['Engineering']
#number of values in x and y are to be same, as they're mapped to each other


#prepare OUTPUT FILE
output_file('Bachelors.html')

#create figure object
f=figure(plot_width=500,plot_height=400)#, tools='pan')#example of attributes of FIGURE


f.title.text="DATA LOOKS MUCH COOLER"
f.title.text_color="RED"
f.title.text_font="times"
f.title.text_font_style="italic"
f.xaxis.minor_tick_line_color=None
f.yaxis.minor_tick_line_color=None
f.xaxis.axis_label="Year"
f.yaxis.axis_label="Engineering"


#create line plot
f.line(x,y)

show(f)
```

```python
import bokeh
import pandas
from bokeh.plotting import figure
from bokeh.io import output_file,show

#prepare data
df=pandas.read_excel("http://pythonhow.com/data/verlegenhuken.xlsx")
x=df['Temperature']/10
y=df['Pressure']/10

#prepare output file
output_file('Weather_Station.html')

#creating figure
f=figure(plot_width=500,plot_height=400)

f.title.text="Temperature and Pressure Graph"
f.title.text_color="red"
f.yaxis.minor_tick_line_color = "red"
f.xaxis.minor_tick_line_color = "red"
f.xaxis.axis_label="Temperature"
f.yaxis.axis_label="Pressure"

f.circle(x,y,size=0.5)
show(f)
```

```python
import bokeh
import pandas
from bokeh.plotting import figure
from bokeh.io import output_file,show

#preparing data
df=pandas.read_csv('table.csv',parse_dates=["Date"])
x=df['Date']
y=df['Close']


#prepare output file
output_file('Table_data.html')

#creating fiure
f=figure(width=500,height=250,x_axis_type="datetime")

f.title.text="Year and Close"
f.title.text_color="red"
f.yaxis.minor_tick_line_color = "red"
f.xaxis.minor_tick_line_color = "red"
f.xaxis.axis_label="Year"
f.yaxis.axis_label="Close"




f.line(x,y,color="orange",alpha=.9,line_width=5) #alpha is for transparency between 0 and 1
f.circle(x,y,color="red",size=1)

show(f)
```
